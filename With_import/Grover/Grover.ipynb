Qiskit

```python
from qiskit import QuantumCircuit, QuantumRegister
qr = QuantumRegister(3)
qc = QuantumCircuit(qr)

# Initialize
qc.h(qr[0])
qc.h(qr[1])
qc.h(qr[2])

# Oracle
qc.ccz(qr[0], qr[1], qr[2])

# Diffusion
qc.h(qr[0]); qc.h(qr[1]); qc.h(qr[2])
qc.x(qr[0]); qc.x(qr[1]); qc.x(qr[2])

qc.h(qr[2])
qc.ccz(qr[0], qr[1], qr[2])
qc.h(qr[2])

qc.x(qr[0]); qc.x(qr[1]); qc.x(qr[2])
qc.h(qr[0]); qc.h(qr[1]); qc.h(qr[2])
```

Cirq

```python
import cirq
q = cirq.LineQubit.range(3)
circuit = cirq.Circuit()

# Initialize
circuit.append([cirq.H(q[0]), cirq.H(q[1]), cirq.H(q[2])])

# Oracle
circuit.append(cirq.CCZ(q[0], q[1], q[2]))

# Diffusion
circuit.append([cirq.H(q[0]), cirq.H(q[1]), cirq.H(q[2])])
circuit.append([cirq.X(q[0]), cirq.X(q[1]), cirq.X(q[2])])

circuit.append(cirq.H(q[2]))
circuit.append(cirq.CCZ(q[0], q[1], q[2]))
circuit.append(cirq.H(q[2]))

circuit.append([cirq.X(q[0]), cirq.X(q[1]), cirq.X(q[2])])
circuit.append([cirq.H(q[0]), cirq.H(q[1]), cirq.H(q[2])])
```

Pennylane

```python
import pennylane as qml

dev = qml.device("default.qubit", wires=3)

@qml.qnode(dev)
def grover():
    # Initialize
    qml.Hadamard(wires=0)
    qml.Hadamard(wires=1)
    qml.Hadamard(wires=2)

    # Oracle (CCZ equivalent)
    qml.ctrl(qml.PauliZ, control=[0, 1])(wires=2)

    # Diffusion
    qml.Hadamard(wires=0)
    qml.Hadamard(wires=1)
    qml.Hadamard(wires=2)

    qml.PauliX(wires=0)
    qml.PauliX(wires=1)
    qml.PauliX(wires=2)

    qml.Hadamard(wires=2)
    qml.ctrl(qml.PauliZ, control=[0, 1])(wires=2)
    qml.Hadamard(wires=2)

    qml.PauliX(wires=0)
    qml.PauliX(wires=1)
    qml.PauliX(wires=2)

    qml.Hadamard(wires=0)
    qml.Hadamard(wires=1)
    qml.Hadamard(wires=2)

    return qml.state()
```
